# 1. Resumen general del proyecto

El proyecto se desarrolló utilizando información de RRHH proveniente de cuatro fuentes:

- empleados (309 registros con datos de cargo, departamento, contrato y jornada);
- ausencias (360 registros con tipos, días y justificación);
- capacitaciones (412 registros con cursos, horas, estado y notas);
- evaluaciones (515 registros con puntajes de desempeño y competencias por periodo).

A partir de estos datos se construyó un dataset integrado orientado a Machine Learning.

Durante el proceso se realizaron:

- normalización de categorías inconsistentes (departamentos, cargos, tipos de ausencia);
- tratamiento de valores extremos (outlier 380 en puntaje, 112M en salario);
- imputación de nulos con estrategia de mediana;
- creación de variables derivadas de RRHH;
- transformaciones logarítmicas para variables sesgadas;
- construcción de indicadores operacionales de riesgo.

El dataset final permitió representar correctamente:

- historial de ausencias de cada empleado;
- participación y rendimiento en capacitaciones;
- nivel de competencias técnicas y blandas;
- perfil organizacional (cargo, departamento, contrato, jornada);
- variables de riesgo de rotación y desempeño.

Esto permitió construir modelos robustos y coherentes desde una perspectiva de gestión de personas.

# 5. Optimización de hiperparámetros

Se aplicaron tres técnicas distintas de optimización en notebook 04:

- GridSearchCV — búsqueda exhaustiva en grilla fija
- RandomizedSearchCV — muestreo aleatorio desde distribuciones estadísticas
- Optuna — optimización bayesiana con aprendizaje de trials anteriores

Adicionalmente, en notebook 03 secciones 9.6-9.8 se entrenaron modelos avanzados (XGBoost, LightGBM) con Optuna y ensambles (Voting, Stacking).

Cada estrategia fue evaluada con validación cruzada sobre el conjunto de entrenamiento, reservando el test set únicamente para evaluación final.

---

## Comparación de estrategias — CLASIFICACIÓN

| Estrategia | Trials | CV Folds | F1-CV (train) | F1 Test | Modelo |
|---|---|---|---|---|---|
| Baseline Random Forest | — | — | 0.4877 | 0.6419 | RF n_est=200 |
| GridSearchCV | 72 comb | 5 | 0.7011 | 0.6467 | RF optimizado |
| RandomizedSearchCV | 30 aleatorios | 5 | 0.712 | 0.6463 | RF optimizado |
| Optuna (RF) | 30 bayesianos | 5 | 0.7117 | 0.6503 | RF optimizado |
| **XGBoost (Optuna)** | **80 trials** | **5** | **0.7174** | **0.7174** | XGB |
| **LightGBM (Optuna)** | **80 trials** | **5** | **0.6979** | **0.6979** | LGBM |
| **VotingClassifier** | — | — | — | **0.7052** | RF+GB+XGB+LGB |
| **Stacking (meta=LR)** | — | **5** | — | **0.7297** | RF+GB+XGB+LGB+LR |

**Interpretación**: 
- El baseline Random Forest (F1-test=0.6419) mejora con optimización a ~0.65 (Optuna RF).
- XGBoost (F1-test=0.7174) supera a Random Forest optimizado.
- **Stacking Ensemble logró el mejor F1=0.7297**, combinando RF, GB, XGBoost y LightGBM con meta-learner Logistic Regression.
- Aunque XGBoost y LightGBM tienen F1-CV ~71-70%, en test generalizan mejor, mostrando validación cruzada correcta sin data leakage.

# 7. Aprendizaje no supervisado

El análisis de clustering en notebook 05 permitió identificar grupos naturales dentro de la población de empleados, **sin usar la variable objetivo supervisada `puntaje_desempeno`**.

Se utilizaron:

- **PCA**: Reducción a 2 dimensiones, explicando 55.65% de la varianza (PC1: 33.6% capacitación, PC2: 22.0% ausencias)
- **KMeans**: Segmentación automática de perfiles
- **Método del codo**: Estimar número óptimo de clusters
- **Silhouette Score**: Validar calidad de clusters (K=3 máximo, pero K=4 seleccionado por utilidad operacional)
- **Davies-Bouldin Index**: Separación entre clusters (DB<1.5 indica buenos clusters)
- **Calinski-Harabasz Index**: Compacidad (CH>30 indica clusters bien definidos)
- **Robustez**: Validación con múltiples random_state (estabilidad <5% cambios)

# 9. Limitaciones del proyecto

Aunque los resultados obtenidos fueron positivos, existen algunas limitaciones importantes que deben considerarse.

## Limitaciones identificadas

- El dataset fue generado con fines académicos y no corresponde a una organización real completa.
- Los datos de ausencias, capacitaciones y evaluaciones tienen cobertura parcial: no todos los empleados tienen registros en todas las tablas.
- El puntaje de desempeño presentó un outlier extremo (valor 380) que requirió tratamiento especial y puede distorsionar el análisis si no se controla correctamente.
- Las variables derivadas fueron construidas a partir de lógica de RRHH definida manualmente, lo que implica criterios subjetivos de diseño.
- No se dispone de información temporal avanzada ni datos longitudinales de largo plazo por empleado.

A pesar de estas limitaciones, el proyecto logra representar adecuadamente una problemática compleja de gestión de personas.
